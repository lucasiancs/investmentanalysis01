```python
#importando bibliotecas
import pandas as pd #leitura de bases de dados
import matplotlib.pyplot as plt #visualização dos dados
import numpy as np #operações algébricas
import pandas_datareader.data as web #fonte de extração de dados
import datetime #formato das datas
%matplotlib inline
```

```python
dia = int(input('Selecione o dia de início:'))
mes = int(input('Selecione o mês de início:'))
ano = int(input('Selecione o ano de início:'))
```

```python
dia1 = int(input('Selecione o dia de fim:'))
mes1 = int(input('Selecione o mês de fim:'))
ano1 = int(input('Selecione o ano de fim:'))
```

```python
start = datetime.datetime(ano,mes,dia)
end = datetime.datetime(ano1,mes1,dia1)
```

```python
ativo = str(input('Digite o código ativo, ex: PETR4.SA'))
```

```python
ativo_dados = web.DataReader(ativo,'yahoo',start,end) #extraindo dados do ativo
dolar_dados = web.DataReader('DEXBZUS','fred',start,end) #extraindo dólar
bovespa_dados = web.DataReader('^BVSP','yahoo',start,end) #extraindo índice bovespa
```

```python
comp, ax = plt.subplots(figsize=(12,6))
ax.plot(bovespa_dados.index,bovespa_dados['Adj Close'],c='green',label='BOVESPA')

ax2 = ax.twinx()
ax2.plot(ativo_dados.index,ativo_dados['Adj Close'],c='red',label=ativo)
ax2.plot(dolar_dados.index,dolar_dados['DEXBZUS'],c='blue',label='Dólar')


ax.set_xlabel('Date')
ax.set_ylabel('Adj Close BOVESPA')
ax2.set_ylabel('Adj Close Ativo')
ax.set_title(f'Comparando: {ativo}, Dólar e o Indíce BOVESPA')

ax.yaxis.grid(True)
comp.legend()
plt.show()
```
